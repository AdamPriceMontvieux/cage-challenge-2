Option Critic Model

```python
from ray.rllib.models.tf import TFModelV2
import tensorflow as tf

class OptionCriticModel(TFModelV2):  

    def __init__(self, obs_space, action_space, num_outputs, model_config, name):

        super(OptionCriticModel, self).__init__(
            obs_space, action_space, num_outputs, model_config, name)
        
        self.num_options = model_config["attention_num_heads"]
        self.num_actions = action_space.n 
        self.current_option = tf.convert_to_tensor([0], dtype=tf.int32)
        self._value_out = 0
        self.temp = 1.0
        # Construct Input
        inputs = tf.keras.layers.Input(shape=obs_space.shape, name="observations")  
        # Construct Core
        layer_1 = tf.keras.layers.Dense(512,name="layer1", activation=tf.nn.tanh)(inputs)
        layer_2 = tf.keras.layers.Dense(512, name="layer2", activation=tf.nn.tanh)(layer_1)
        # Construct Option Layers
        options = []
        for i in range(self.num_options):
            o0_0 = tf.keras.layers.Dense(256, name="o"+str(i)+"_0", activation=tf.nn.relu)(layer_2)
            o0_1 = tf.keras.layers.Dense(self.num_actions, name="o"+str(i)+"_1", activation='linear')(o0_0)
            options.append(o0_1)
        # Construct Option Critic Layer
        oc = tf.keras.layers.Dense(self.num_options, name="option_critic", activation='linear')(layer_2)
        # Construct Termination Layer
        t = tf.keras.layers.Dense(self.num_options, name="temination", activation=tf.nn.sigmoid)(layer_2)
        # Construct Value Layer
        V = tf.keras.layers.Dense(1, activation='linear', name='value')(layer_2)

        self.base_model = tf.keras.Model(inputs, [options, oc, t, V])  

    # Implement the core forward method.
    def forward(self, input_dict, state, seq_lens):
        model_out = self.base_model([input_dict["obs"]])
        self._value_out = model_out[3][0]
        return model_out, state
```
